### Fixing "normalized-losses" column

```python
# "?" is an object so it cannot be converted to median.
# We have to first replace it with 0 then convert the column into 
# integer and then replace 0 with median value.
```

```python
df["normalized-losses"] = df["normalized-losses"].replace(to_replace ="?",
                 value =0)
```

```python
df["normalized-losses"].astype(int)
```

```python
df["normalized-losses"].median()
```

```python
df["normalized-losses"].replace({0:df["normalized-losses"].median()},inplace=True)
```

```python
df.head()
```

```python
df=df[df["price"]!="?"] #drop raw where price of value is "?". Because price is the targeted feature
```
